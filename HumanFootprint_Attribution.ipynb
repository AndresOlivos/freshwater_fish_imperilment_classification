## Human Footprint Attribution

```python
import pandas as pd
import exactextract
```

```python
#input_ranges = r"C:\Users\olivoshj\Desktop\ArcGIS_Projects\Dam_Project_Aux\Ranges_SQLite.gpkg"
input_ranges = r"D:\Andres\Dam_Project_D\scratch\sqlite.gpkg"
```

```python
# Human Footprint data downloaded from: https://doi.org/10.5061/dryad.052q5
hfp_1993 = r"R:\FWL\Arismendi-Lab\Andres\Gilbert_Freshwater_Fish_Analysis\Environmental_Data\Human_Footprint_Dryad\Dryad_HFP\HFP1993.tif"
hfp_2009 = r"R:\FWL\Arismendi-Lab\Andres\Gilbert_Freshwater_Fish_Analysis\Environmental_Data\Human_Footprint_Dryad\Dryad_HFP\HFP2009.tif"
```

```python
for hfp in [hfp_1993,hfp_2009]:
    arcpy.management.ProjectRaster(
        in_raster=hfp,
        out_raster=fr"{hfp}_wgs84.tif",
        out_coor_system='GEOGCS["GCS_WGS_1984",DATUM["D_WGS_1984",SPHEROID["WGS_1984",6378137.0,298.257223563]],PRIMEM["Greenwich",0.0],UNIT["Degree",0.0174532925199433]]',
    )
```

```python
df_1993 = exactextract.exact_extract(rast=fr"{hfp_1993}_wgs84.tif",
                                     vec=input_ranges,
                                     ops="hfp_1993=mean",
                                     include_cols=["sci_name"],
                                     include_geom=False,
                                     output='pandas',
                                     progress=True
                                    )
```

```python
df_2009 = exactextract.exact_extract(rast=fr"{hfp_2009}_wgs84.tif",
                                     vec=input_ranges,
                                     ops="hfp_2009=mean",
                                     include_cols=["sci_name"],
                                     include_geom=False,
                                     output='pandas',
                                     progress=True
                                    )
```

```python
hfp_df = df_1993.merge(df_2009,on='sci_name')
```

```python
hfp_df.to_csv(r"R:\FWL\Arismendi-Lab\Andres\Gilbert_Freshwater_Fish_Analysis\Revised_Analysis_NatureCommunications\HFP_Attributes_Global_Grid.csv", 
             index=False)
```

```python
for hfp in [hfp_1993,hfp_2009]:
    arcpy.management.Delete(fr"{hfp}_wgs84.tif")
```
