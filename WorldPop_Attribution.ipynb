## WorldPop Attribution

```python
import pandas as pd
import exactextract
```

```python
input_ranges = r"C:\Users\olivoshj\Desktop\ArcGIS_Projects\Dam_Project_Aux\Ranges_SQLite.gpkg\Extant_Native_Ranges_Clipped"
```

```python
# WorldPop data downloaded from https://hub.worldpop.org/geodata/listing?id=64
wp_2000 = r"R:\FWL\Arismendi-Lab\Andres\Gilbert_Freshwater_Fish_Analysis\Revised_Analysis_NatureCommunications\Input_datasets\Socioeconomic_datasets\WorldPop\ppp_2000_1km_Aggregated.tif"
wp_2020 = r"R:\FWL\Arismendi-Lab\Andres\Gilbert_Freshwater_Fish_Analysis\Revised_Analysis_NatureCommunications\Input_datasets\Socioeconomic_datasets\WorldPop\ppp_2020_1km_Aggregated.tif"
```

```python
df_2000 = exactextract.exact_extract(rast=wp_2000,
                                     vec=input_ranges,
                                     ops="EarlyPop=sum",
                                     include_cols=["sci_name","Clipped_AreaSqk"],
                                     include_geom=False,
                                     output='pandas',
                                     progress=True
                                    )
```

```python
df_2020 = exactextract.exact_extract(rast=wp_2020,
                                     vec=input_ranges,
                                     ops="LatePop=sum",
                                     include_cols=["sci_name"],
                                     include_geom=False,
                                     output='pandas',
                                     progress=True
                                    )
```

```python
wp_df = df_2000.merge(df_2020,on='sci_name')
```

```python
# calculate variables of interest
wp_df["EarlyPD"] = wp_df["EarlyPop"] / wp_df["Clipped_AreaSqk"] # pop density 2000
wp_df["LatePD"] = wp_df["LatePop"] / wp_df["Clipped_AreaSqk"] # pop density 2020
wp_df["PDpercent"] = (wp_df["LatePD"] - wp_df["EarlyPD"]) / wp_df["EarlyPD"] * 100 / 20 # average percent increase per year
wp_df["PDslope"] = (wp_df["LatePD"] - wp_df["EarlyPD"]) / 20 # average increase per year
```

```python
wp_df.to_csv(r"R:\FWL\Arismendi-Lab\Andres\Gilbert_Freshwater_Fish_Analysis\Revised_Analysis_NatureCommunications\Input_datasets\WorldPop_Attributes.csv", 
             index=False)
```
